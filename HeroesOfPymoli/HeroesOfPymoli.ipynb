```python
#Import pandas
import pandas as pd
```

```python
#Open and read csv file
filePath='purchase_data.csv'

p=pd.read_csv(filePath)
p.head()
```

```python
#Player Count
#displaying total number of players
    
uniquePlayers=len(p['SN'].unique())

playerCount=pd.DataFrame({
    'Total Players':[uniquePlayers]
})
playerCount
```

```python
#Purchasing Analysis (Total)

#Calculating the values to display in the Data Frame
pa_uniqueItems=len(p['Item ID'].unique()) #Unique Items
pa_avgPrice=p['Price'].mean() #Average Price
pa_totalPurchases=p['Purchase ID'].count() #Number of Purchases
pa_totalRevenue=p['Price'].sum() #Total Revenue

#Create Purchasing Analysis (Total) Data Frame
purchasingAnalysis=pd.DataFrame({
    'Number of Unique Items':[pa_uniqueItems],
    'Average Price':[pa_avgPrice],
    'Number of Purchases':[pa_totalPurchases],
    'Total Revenue':[pa_totalRevenue]
})

#Formating Average Price and Total Revenue columns.
purchasingAnalysis['Average Price']=purchasingAnalysis['Average Price'].map("${:.2f}".format)
purchasingAnalysis['Total Revenue']=purchasingAnalysis['Total Revenue'].map("${:,.2f}".format)

#Displaying the table
purchasingAnalysis
```

```python
#Gender Demographics

#Creating a copy of the main data frame
p_copy=p.copy()

#Deleting duplicates to keep only unique users
uniquep=p_copy.drop_duplicates('SN')

#Calculating the values to display in the Data Frame
gd_genderCounts=uniquep["Gender"].value_counts() #counts per gender
gd_genderpercents=gd_genderCounts/uniquePlayers*100 #percentage of total

#Creating Data Frame
genderDemographics=pd.DataFrame({
    'Total Count':gd_genderCounts,
    'Percentage of Players':gd_genderpercents
})

#Formating Percentage of Players' column
genderDemographics['Percentage of Players']=genderDemographics['Percentage of Players'].map("{:.2f}%".format)

#Displaying Table
genderDemographics
```

```python
#Purchasing Analysis (Gender)

#Calculating the values Purchase Count, Average Purchase Price and Total Purchase Value to display in the Data Frame
pag_1=p.groupby("Gender") #Grouping by gender the main table with all the purchases
pag_purchaseCount=pag_1['Purchase ID'].count() #Getting the count of purchases per each gender
pag_averagePrice=pag_1["Price"].mean() #average purchase price per gender
pag_totalPurchaseValue=pag_1['Price'].sum() #total purchase value per gender

pag_2=uniquep.groupby("Gender") #Grouping by Gender the table with unique users
pag_uniqueCountGender=pag_2['SN'].count() #Counting unique users per gender

#Calculating the Avg Total Purchase per Person
pag_averageTotalPerson=pag_totalPurchaseValue/pag_uniqueCountGender

#Creating Data Frame
purchasingAnalysisGender=pd.DataFrame({
    'Purchase Count':pag_purchaseCount,
    'Average Purchase Price':pag_averagePrice,
    'Total Purchase Value':pag_totalPurchaseValue,
    'Avg Total Purchase per Person':pag_averageTotalPerson
})

#Formating all the columns except the first one
for index,x in enumerate(purchasingAnalysisGender):
    if index!=0:
        purchasingAnalysisGender[x]=purchasingAnalysisGender[x].map("${:,.2f}".format)

#Displaying Table
purchasingAnalysisGender
```

```python
#Age Demographics

#Creating a copy of the unique users' data frame
ad=uniquep.copy()

#calculating maximum age for teh bins
ad_max_age=ad['Age'].max()

#List of labels for the bins
labels=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

#list of bins
bins=[0,9,14,19,24,29,34,39,ad_max_age]

#creating the age bins/categories
ad["Age Ranges"] = pd.cut(ad["Age"], bins, labels=labels,include_lowest=True)

#Calculating the values to display in the Data Frame
ad_TotalCount=ad['Age Ranges'].value_counts(sort=False) #count of users in each age bin
ad_agePercentPlayers=(ad_TotalCount/uniquePlayers)*100 #percentage of total

#Creating Data frame
ageDemographic=pd.DataFrame({
    'Total Count':ad_TotalCount,
    'Percentage of Players':ad_agePercentPlayers    
})


#Formating the Percentage of Players' column
ageDemographic['Percentage of Players']=ageDemographic['Percentage of Players'].map('{:.2f}%'.format)

#Displaying table
ageDemographic
```

```python
#Purchasing Analysis (Age)

#Creating a copy of the previous table already with the Age Ranges
paa=p.copy()

#Creating the age ranges but no with the main table instead of the unique users table
paa['Age Ranges']=pd.cut(paa["Age"], bins, labels=labels,include_lowest=True)

#Grouping by Age Groups
paa_group=paa.groupby('Age Ranges')

#Calculating the values to display in the Data Frame
paa_agePurchaseCount=paa_group['Purchase ID'].count() #Number of purchases by age range
paa_avgPurchasePrice=paa_group['Price'].mean() #Avg purchase price by age range
paa_totalPurchaseValue=paa_group['Price'].sum() # Total purchase value by age range
paa_avgTotalPurchasePerson=paa_totalPurchaseValue/ad_TotalCount #Avg total purchase per person per age range

#Creating Data frame
purchasingAnalysis_Age=pd.DataFrame({
    'Purchase Count':paa_agePurchaseCount,
    'Average Purchase Price':paa_avgPurchasePrice,
    'Total Purchase Value':paa_totalPurchaseValue,
    'Avg Total Purchase per Person':paa_avgTotalPurchasePerson
})

#Formating all the columns except the first one
for index,x in enumerate(purchasingAnalysis_Age):
    if index!=0:
        purchasingAnalysis_Age[x]=purchasingAnalysis_Age[x].map("${:,.2f}".format)

#Displaying table
purchasingAnalysis_Age
```

```python
#Top Spenders

#Creating a copy of the main table
ts=p.copy()

#Grouping by user
ts_group=ts.groupby('SN')

#Calculating the values to display in the Data Frame
ts_purchaseCount=ts_group['Item ID'].count() #Number of purchases per user
ts_avgPurchasePrice=ts_group['Price'].mean() #Avg purchase price per user
ts_totalPurchaseValue=ts_group['Price'].sum() #Total purchase value per user

#Creating Data frame
topSpenders=pd.DataFrame({
    'Purchase Count':ts_purchaseCount,
    'Average Purchase Price':ts_avgPurchasePrice,
    'Total Purchase Value':ts_totalPurchaseValue
})

#Creating a new data frame with only the top 5 values sorted by Total Purchase value
topSpendersSummary=topSpenders.sort_values(by=['Total Purchase Value'],ascending=False).head()

#Formating all the columns except the first one
for index,x in enumerate(topSpenders):
    if index!=0:
        topSpendersSummary[x]=topSpendersSummary[x].map("${:,.2f}".format)

#Displaying table
topSpendersSummary
```

```python
#Most Popular Items

#Creating a copy of the main table
mpi=p.copy()

#Grouping by item ID and Item Name
mpi_group=mpi.groupby(['Item ID','Item Name'])

#Calculating the values to display in the Data Frame
mpi_purchaseCount=mpi_group['Purchase ID'].count() #Count of purchases per item
mpi_itemPrice=mpi_group['Price'].max() #Price of the item
mpi_totalPurchaseValue=mpi_group['Price'].sum() #Total ammount purchase per item

#Creating Data frame
mostPopularItems=pd.DataFrame({
    'Purchase Count':mpi_purchaseCount,
    'Item Price':mpi_itemPrice,
    'Total Purchase Value':mpi_totalPurchaseValue
})

#Creating a new data frame with only the top 5 values sorted by Purchase Count
mostPopularItems_summary=mostPopularItems.sort_values(by=['Purchase Count'],ascending=False).head()

#Formating all the columns except the first one
for index,x in enumerate(mostPopularItems_summary):
    if index!=0:
        mostPopularItems_summary[x]=mostPopularItems_summary[x].map("${:,.2f}".format)

#Displaying table
mostPopularItems_summary
```

```python
#Most Profitable Items

#Creating Data Frame from the previously created table but only with the top 5 values sorted by Total Purchase Value
mostProfitableItems=mostPopularItems.sort_values(by=['Total Purchase Value'],ascending=False).head()

#Formating all the columns except the first one
for index, x in enumerate(mostProfitableItems):
    if index!=0:
        mostProfitableItems[x]=mostProfitableItems[x].map('${:.2f}'.format)

#Displaying table
mostProfitableItems
```
